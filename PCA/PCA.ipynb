# PCA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from data_helper import XY_from_df
```

```python
TRAIN_PATH = "dataset/train.csv"
df_train = pd.read_csv(TRAIN_PATH)
X_train, Y_train = XY_from_df(df_train)
```

```python
colors = []

for idx, y in enumerate(Y_train):
    if y == 0: colors.append('r')
    if y == 1: colors.append('b')
```

```python
pca = PCA(n_components=3)
pca.fit(X_train)
```

```python
pca.explained_variance_ratio_
```

```python
X_new = pca.transform(X_train)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_new[:, 0], X_new[:, 1], X_new[:, 2], s=70, c=colors, alpha=.5)
```
